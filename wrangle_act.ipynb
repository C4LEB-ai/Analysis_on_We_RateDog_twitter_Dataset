# Define
#### From the twitter_archive_clean
- Drop retweeted_status_timestamp", "retweeted_status_user_id","retweeted_status_id", "in_reply_to_user_id", "in_reply_to_status_id"
- convert the four columns containing the stages of dog into one single column (doggo, floofer, pupper, puppo)
- Drop the four columns and use the newly created column (stage)
- corect the name errors and replace none with Nan

```python
twitter_archive_clean.drop(["retweeted_status_timestamp", "retweeted_status_user_id","retweeted_status_id",
                           "in_reply_to_user_id", "in_reply_to_status_id"], axis = 1, inplace = True)
```

```python
# Replacing the None values with NaN in the doggo, floofer, pupper and puppo column of the twitter_archive_clean
stages = ['doggo', 'floofer', 'pupper', 'puppo']
for stage in stages:
    twitter_archive_clean[stage] = twitter_archive_clean[stage].apply(lambda x: np.NaN if x == 'None' else x)
    
## Converting the four stages into one named "stages"
twitter_archive_clean['stage'] = twitter_archive_clean[['doggo', 'floofer', 'pupper', 'puppo']].astype(str).sum(1)

##replacing the 'nan', "nannannan" values for the values concatenated above with empy space
twitter_archive_clean['stage'] = twitter_archive_clean['stage'].apply(lambda x: x.replace('nan', ''))

# replacing the empty space above wit Nan 
twitter_archive_clean['stage'] = twitter_archive_clean['stage'].apply(lambda x: np.NaN if x == '' else x)
```

```python
# now we need to seperate the values that was merged above
twitter_archive_clean['stage'].replace({'doggopupper':'doggo pupper', 
                                'doggofloofer':'doggo floofer',
                                'doggopuppo':'doggo puppo',
                                '':None}, inplace=True)
```

```python
## we now drop the four stages (doggo	floofer	pupper	puppo) since we have created a single column for that
twitter_archive_clean.drop(['doggo', 'floofer', 'pupper', 'puppo'], axis=1, inplace=True)
```

```python
# now we need to seperate the values that was merged above
twitter_archive_clean['name'].replace({'a':'NaN', 
                                'None':'NaN',
                                'Nan':'NaN'}, inplace=True)
```
